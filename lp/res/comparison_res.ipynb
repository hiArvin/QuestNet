```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
f=open('no_slice.txt','r')
l1=f.readline()
opt_time = []
for r in l1.split():
    opt_time.append(float(r))
l2=f.readline()
modelling_time=  []
for m in l2.split():
    modelling_time.append(float(m))

l3=f.readline()
fail_num=  []
for fn in l3.split():
    fail_num.append(float(fn))
print(sum(opt_time))
print(sum(modelling_time))
print(sum(fail_num)/1500)
```

```python
plt.plot(opt_time)
plt.xlabel("Epoches")
# plt.ylabel("Optimal time")
plt.title("Optimizing time")
plt.show()
```

```python
plt.plot(modelling_time)
plt.xlabel("Epoches")
# plt.ylabel("Failure rate")
plt.title("Modelling time")
plt.show()
```

```python
plt.plot(np.array(fail_num)/30)
plt.xlabel("Epoches")
# plt.ylabel("Failure rate")
plt.title("Failure Rate")
plt.show()
```
