```python
import matplotlib.pyplot as plt
```

```python
f=open('res/p=1.txt','r')
l1=f.readline()
runtime = []
for r in l1.split():
    runtime.append(float(r))

l2=f.readline()
fail_rate=  []
for f in l2.split():
    fail_rate.append(float(f))
print(runtime)
print(fail_rate)
```

```python
x_dim= list(range(10,110,10))
x_dim
```

```python
plt.plot(x_dim,fail_rate)
plt.xlabel("Number of quests")
plt.ylabel("Failure rate")
plt.title('Number of paths:1')
plt.show()
```

```python
plt.plot(x_dim,runtime)
plt.xlabel("Number of quests")
plt.ylabel("Runtime")
plt.title('Number of paths:1')
plt.show()
```
